# Nomor 3
## Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.
Untuk menampilkannya, saya menggunakan variabel untuk menampung isi dari jumlah record, jumlah attribute, nama attribute, dan jumlah attribute yang bertipe data angka.
- Untuk menampilkan jumlah record dan attribute saya  menggunakan fungsi **shape()**
- Untuk menampilkan nama attribute saya menggunakan **list(df.columns)** yang berguna untuk menampilkan nama kolom dalam bentuk list
- Untuk menampilkan Attribute yang bertipe angka saya menggunakan fungsi **len()** untuk mengambil nilai dari tipe data yang telah diseleksi menggunakan fungsi **select_dtypes(include=['number']).columns)**
- Lalu, saya memanggil variabel **info_dict()** dengan perulangan agar data yang ada tersusun dengan baik

```python
info_dict = {
    "Jumlah Record": df.shape[0],
    "Jumlah Attribute": df.shape[1],
    "Nama Attribute": list(df.columns),
    "Jumlah Attribute Tipe Angka": len(df.select_dtypes(include=['number']).columns)
}

i = 1
for key, value in info_dict.items():
    print("==========================================")
    if key == "Nama Attribute":
        print(f"||{i} | {key} : \n{value}")
    else:
        print(f"||{i} | {key} : {value}")    
    i+=1
    
```

## Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice
### NIM Akhir = 077
Untuk menampilkan nilai yang ada dari baris ke 77 sampai dengan akhir menggunakan dua metode slicing : 
- loc[]
  Saya menggunakan metode ini dengan cara memanggil baris terlebih dahulu dan diikuti dengan df.columns[nomor kolom:]


```python
df.loc[77:, df.columns[3:]]
```

- iloc[].
  Saya menggunakan metode ini dengan cara memanggil kolom dan baris yang diinginkan. Contoh : [77:, 3:]

```python
df.iloc[77:, 3:]
```
