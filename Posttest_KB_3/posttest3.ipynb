# Import package yang akan digunakan

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

# Membuat variabel df untuk menampung dataset yang akan digunakan

```python
df = pd.read_csv("FAKE NEWS DATASET.csv")
```

## Membuat variabel Label untuk menampung jumlah value dari attribute Label

```python
Label = df["Label"].value_counts()
```

# Pie Chart
Menggunakan plt.pie() untuk menampilkan pie chart dengan memanggil variabel *Label* lalu mengambil nama labels dengan menggunakan fungsi .index untuk mengetahui persentase tersebut milik siapa. Pie chart ini dimulai dari sudut 90 derajat menggunakan fungsi **startangle = 90** dan mengambil dua angka di belakang koma dengan fungsi **autopct = "%.2f%%"**.
Menampilkan judul Pie Chart dengan menggunakan fungsi **plt.title("Persentase Berita Asli dan Palsu")**

```python

plt.pie(Label, labels = Label.index, startangle = 90, autopct = "%.2f%%")
plt.title("Persentase Berita Asli dan Palsu")
plt.show()
```

Pada pie chart di atas, dapat diketahui bahwa pada dataset tersebut memiliki tiga kategori, yaitu **Fake**, **Real**, dan **1**. Pada kategori **Fake** terdapat **61.07 %** berita palsu yang beredar, sedangkan pada kategori **Real** terdapat **38.89 %** berita yang asli kebenarannya, yang terakhir ada kategori **1** dengan persentase sebanyak **0.05 %**

# Heatmap
Untuk membuat heatmap, saya menggunakan fungsi **.heatmap()** yang dipanggil bersamaan dengan **df.corr()**. Fungsi dari **df.corr()** ini untuk menunjukkan korelasi antara attribute yang bertipe numerik. fungsi **cmap** digunakan untuk merubah warna, dan yang terakhir adalah **annot = True**, fungsi ini digunakan untuk menaruh angka atau anotasi di dalam kotak heatmap tersebut.

```python
sns.heatmap(df.corr(), cmap = "rainbow", annot = True)
plt.title("Korelasi Antara Attribute Bertipe Numerik")
plt.show()
```

Pada heatmap di atas, diketahui bahwa hanya ada satu attribute yang bertipe numerik yaitu **HasImage**. Sehingga dapat diketahui bahwa korelasinya hanya dengan attributenya sendiri.

# Diagram Batang

```python
x= df["Language"].value_counts().index
y = df["Language"].value_counts()
plt.bar(x, y , color=["blue","yellow","green"])
plt.title("Jumlah Berita Berdasarkan Bahasa")
plt.xlabel("Bahasa")
plt.ylabel("Jumlah")
plt.show()
```

Dari data diagram batang di atas, bahasa **Inggris** merupakan bahasa yang sering digunakan dalam penulisan berita dengan jumlah sebanyak lebih dari **2.000** berita, lalu di urutan terbanyak kedua ada berita yang berbahasa **Jerman** dengan jumlah berita sebanyak kurang dari **250** berita, sedangkan kategori bahasa lain seperti **Ignore, French, Hate,** dan **Spanish** memiliki data yang sedikit yaitu kurang dari **10** berita.

# Histogram
Untuk membuat histogram, saya menggunakan fungsi plt.hist(). Histogram ini dibuat untuk mengetahui frekuensi distribusi data dari attribute **Type.**Fungsi **plt.title()** digunakan untuk menaruh judul. Fungsi **xlabel()** digunakan untuk memberikan label pada sumbu x dan **ylabel()** digunakan untuk menaruh label pada sumbu y.

```python
plt.hist(df["HasImage"], bins = 15, color='skyblue')
plt.title("Histogram Attribute Has Image")
plt.xlabel("Nilai dari Attribute")
plt.ylabel("Frekuensi Kemunculan")
plt.show()
```

Pada data di atas, dapat diketahui bahwa frekuensi kemunculan data yang bernilai **1.0** merupakan data yang memiliki frekuensi kemunculan yang paling tinggi, yaitu sebanyak lebih dari **1.600 kali**, sedangkan frekuensi data yang bernilai **0** hingga kurang dari **0.1** adalah lebih dari **400 kali**.
